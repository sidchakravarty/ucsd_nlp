<h3 style = "color:lightgreen">1.1.2 Introduction to Regular Expressions</h3>

```python
text_to_search = '''

    abcdefghikjklmnopqrstuvwxyz
    ABCDEFGHIJKLMNOPQRSTUVWXYZ
    1234567890

    Ha HaHa

    MetaCharacters  (Need to be escaped):
    . ^ $ * + ? { } [ ] \ | ( )

    ash@ashpahwa.com

    321-312-3123
    332.121.3133
    123*143*2344

    Mr.    Schafer
    Mr Smith
    Ms Davis
    Mrs.    Robinson
    Mr. T
'''

sentence = 'Start a sentence and then bring it to an end.'
```

```python
pattern = re.compile(r'abc')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
print(text_to_search[match.start():match.end()])
```

```python
# Match all characters
pattern = re.compile(r'.').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Find periods (.)

pattern = re.compile(r'\.').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Search for a string that includes a period, such as an email address.

pattern = re.compile(r'@').finditer(text_to_search)

for match in pattern:
    print(match)
    print(text_to_search[match.start():match.end()])
```

```python
# Search for a string that includes a period, such as an email address.

pattern = re.compile(r'ashpahwa\.com').finditer(text_to_search)

for match in pattern:
    print(match)
    print(text_to_search[match.start():match.end()])
```

```python
# Search for all digits in a string.

pattern = re.compile(r'\d').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Search for all characters, digits.

pattern = re.compile(r'\w').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Search for all whitespace characters.

pattern = re.compile(r'\s').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Search for a word boundary in a string.

pattern = re.compile(r'\bHa').finditer(text_to_search)

for match in pattern:
    print(match)
# Search for a word boundary in a string.
```

```python
# Search for a string that starts with a word

pattern = re.compile(r'^Start').finditer(sentence)

for match in pattern:
    print(match)
```

```python
# Search for a string that ends with a word

pattern = re.compile(r'end.$').finditer(sentence)

for match in pattern:
    print(match)
```

```python
# Search for string that ends with the letter a

pattern = re.compile(r'a$').finditer(sentence)

for match in pattern:
    print(match)
```

```python
# Search for a phone number in a string

pattern = re.compile(r'\d\d\d.\d\d\d.\d\d\d\d').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
with open(r'H:\AI\NLP\UCSD\ucsd_nlp\0. Data\Regular Expression Sample.txt', encoding='utf-8') as f:
    contents = f.read()
```

```python
# Match telephone numbers in a file

pattern = re.compile(r'\d\d\d.\d\d\d.\d\d\d\d').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Match single characters in a file# The square brackets are used to match a single character.
# In this example, we are matching either a '-' or a '.'.

pattern = re.compile(r'\d\d\d[-.]\d\d\d[-.]\d\d\d\d').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Match phone numbers starting with either 800 or 900

pattern = re.compile(r'[89]00[-.]\d\d\d[-.]\d\d\d\d').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Search for a range of numbers in a phone number

pattern = re.compile(r'[1-5]').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Find all small letters in a file

pattern = re.compile(r'[a-z]').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Find all small and capital letters in a file

pattern = re.compile(r'[a-zA-Z]').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Find all non-characters in a file using a caret

pattern = re.compile(r'[^a-zA-Z]').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Match all words that end in at, but not have bat

pattern = re.compile(r'\w+[^b]at\b').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Find phone numbers using quantifiers

pattern = re.compile(r'\d{3}.\d{3}.\d{4}').finditer(contents)

for match in pattern:
    print(match)
```

```python
# Find names that start with Mr.
# The question mark indicates that the period is optional.

pattern = re.compile(r'Mr\.?').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Find names that start with Mr. and a name that has an upper case letter

pattern = re.compile(r'Mr\.?\s[A-Z]').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Find names that start with Mr. and a name that has an upper case letter

pattern = re.compile(r'Mr\.?\s[A-Z]\w+').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Find names that start with Mr. and a name that has an upper case letter

pattern = re.compile(r'Mr\.?\s[A-Z]\w*').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Find names that start with Mr. and a name that has an upper case letter

pattern = re.compile(r'Mr\.?\s*[A-Z]\w*').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Find all names that start with Mr., Ms., or Mrs.
# The parantheses are used to group the characters.

pattern = re.compile(r'M(r|s|rs)\.?\s*[A-Z]\w*').finditer(text_to_search)

for match in pattern:
    print(match)
```

```python
# Matching Email Addresses

emails = '''
CoreyMSchafer@gmail.com
corey.mschafer@university.edu
corey-321-schafer@my-work.com '''

pattern = re.compile(r'[a-zA-Z0-9.-]+@[a-zA-Z-]+\.(com|edu)').finditer(emails)

for match in pattern:
    print(match)
```

```python
# Matching URLs

urls = '''
http://www.google.com
https://www.google.com
http://www.google.com/
https://www.nasa.gov/ '''

pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)').finditer(urls)

for match in pattern:
    print(match)
```
